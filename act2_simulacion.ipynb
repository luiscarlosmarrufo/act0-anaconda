```python
import numpy as np
import matplotlib.pyplot as plt
```

## Weibull

```python
n = 10000
alpha = 10
beta = 2

X = []
for _ in range(n):
    u = np.random.uniform(0,1)
    x = beta * (-np.log(1 - u))**(1.0/alpha)
    X.append(x)
X = np.array(X)

plt.figure()
plt.hist(X, bins=30)
plt.show()

print("media, var =", round(X.mean(), 4), ",", round(X.var(ddof=1), 4))
```

## Box-Muller

```python
n = 10000
m = n // 2
u1 = np.random.uniform(0,1,m)
u2 = np.random.uniform(0,1,m)

r = np.sqrt(-2.0*np.log(u1))
theta = 2.0*np.pi*u2
z1 = r*np.cos(theta)
z2 = r*np.sin(theta)

Z = np.concatenate([z1, z2])
if Z.size < n:  # por si n es impar
    Z = np.append(Z, (np.sqrt(-2*np.log(np.random.uniform())) *
                      np.cos(2*np.pi*np.random.uniform())))
Z = Z[:n]

plt.figure()
plt.hist(Z, bins=30)
plt.show()

print("media, var =", round(Z.mean(), 4), ",", round(Z.var(ddof=1), 4))
```

## Distribucion Geometrica

```python
n = 500
p = 0.25

X = []
for x in range(n):
    u = np.random.uniform(0,1)
    i = 1
    pr = p
    F = pr
    while u >= F:
        pr = (1 - p) * pr
        F += pr
        i += 1
    x = i
    X.append(i)
```

```python
plt.hist(X, bins=range(1, max(X)+2), density=True, rwidth=0.8)
plt.xlabel("x")
plt.ylabel("pmf aproximada")
plt.show()
```
